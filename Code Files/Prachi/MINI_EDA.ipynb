1. Bar Plot

   
Problem Statement:
Compare the average salary across different cities to identify high-paying locations.

```python
import matplotlib.pyplot as plt

# Bar plot: Average salary by city
avg_salary = df.groupby('City')['Salary'].mean()
avg_salary.plot(kind='bar', color='skyblue', figsize=(8, 6))

```

2. Line Plot
   
Problem Statement:
Visualize the trend in employee salaries as age increases.

```python
# Line plot: Salary vs. Age
df.sort_values(by='Age', inplace=True)
plt.figure(figsize=(8, 6))
plt.plot(df['Age'], df['Salary'], marker='o', linestyle='-', color='green')
plt.title('Salary Trend by Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.show()
```

3. Scatter Plot
   
Problem Statement:
Analyze the relationship between age and salary to check for correlation.

```python
# Scatter plot: Age vs. Salary
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'], color='orange')
plt.title('Age vs. Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

4. Histogram
   
Problem Statement:
Understand the distribution of employee ages.

```python
# Histogram: Age distribution
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=10, color='purple', edgecolor='black')
plt.title('Distribution of Employee Ages')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

5. Pie Chart

   
Problem Statement:
Visualize the proportion of employees in each city.

```python
# Pie chart: Employee count by city
city_counts = df['City'].value_counts()
city_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), startangle=90)
plt.title('Employee Distribution by City')
plt.ylabel('')  # Hide y-axis label
plt.show()
```

6. Box Plot

Problem Statement:
Analyze the spread of salaries and identify outliers.

```python
# Box plot: Salary distribution
plt.figure(figsize=(8, 6))
plt.boxplot(df['Salary'].dropna(), vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.show()

```

7. Heatmap (requires additional libraries)

Problem Statement:
Find correlations between numerical features in the dataset

```python
import seaborn as sns

# Heatmap: Correlation matrix
correlation_matrix = df[['Age', 'Salary']].corr()
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

8. Grouped Bar Plot
   
Problem Statement:
Compare average age and salary across cities.

```python
# Grouped bar plot: Age and Salary by city
avg_stats = df.groupby('City')[['Age', 'Salary']].avg()


avg_stats.plot(kind='bar', figsize=(10, 6), color=['blue', 'red'])
plt.title('Average Age and Salary by City')
plt.xlabel('City')
plt.ylabel('Average Values')
plt.xticks(ticks=None,rotation=45)
plt.legend(['Age', 'Salary'])
plt.show()
```
